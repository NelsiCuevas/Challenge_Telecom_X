#📌 Extracción

```python
url = "https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json"
```

```python
import pandas as pd
import requests
import json
import warnings
warnings.filterwarnings('ignore')
```

```python
datos_clientes = requests.get(url)
```

```python
resultado= json.loads(datos_clientes.text)
pd.DataFrame(resultado)
```

```python
df_clientes = pd.json_normalize(resultado)
df_clientes
```

## Diccionario de datos

* **customerID:** número de identificación único de cada cliente
* **Churn:** si el cliente dejó o no la empresa
* **gender:** género (masculino y femenino)
* **SeniorCitizen:** información sobre si un cliente tiene o no una edad igual o mayor a 65 años
* **Partner:** si el cliente tiene o no una pareja
* **Dependents:** si el cliente tiene o no dependientes
* **tenure:** meses de contrato del cliente
* **PhoneService:** suscripción al servicio telefónico
* **MultipleLines:** suscripción a más de una línea telefónica
* **InternetService:** suscripción a un proveedor de internet
* **OnlineSecurity:** suscripción adicional de seguridad en línea
* **OnlineBackup:** suscripción adicional de respaldo en línea
* **DeviceProtection:** suscripción adicional de protección del dispositivo
* **TechSupport:** suscripción adicional de soporte técnico, menor tiempo de espera
* **StreamingTV:** suscripción de televisión por cable
* **StreamingMovies:** suscripción de streaming de películas
* **Contract:** tipo de contrato
* **PaperlessBilling:** si el cliente prefiere recibir la factura en línea
* **PaymentMethod:** forma de pago
* **Charges.Monthly:** total de todos los servicios del cliente por mes
* **Charges.Total:** total gastado por el cliente

#🔧 Transformación

```python
#duplicados
df_clientes.duplicated().sum()
```

```python
#datos unicos
df_clientes.nunique()
```

```python
for col in df_clientes.columns:
  print(f'{col}: {df_clientes[col].nunique()}')
  if df_clientes[col].nunique() < 50:
    print(df_clientes[col].unique())
    print('-' * 50)
```

```python
#nulos
df_clientes.isnull().sum()
```

```python
#celdas vacias
(df_clientes == "").sum()
```

```python
#elimio filas vacias de churn
df_clientes= df_clientes[df_clientes['Churn'].str.strip() != '']
```

```python
#verifico celdas vacias
(df_clientes == "").sum()
```

```python
#celdas con cadenas vacias
(df_clientes == " ").sum()
```

```python
#tranformo a float Charges y elimino vacios
df_clientes["account.Charges.Total"] = pd.to_numeric(df_clientes["account.Charges.Total"], errors="coerce")
df_clientes = df_clientes.dropna(subset=["account.Charges.Total"])
df_clientes
```

```python
#verifico
(df_clientes == " ").sum()
```

```python
df_clientes[df_clientes["account.Charges.Total"].isna()]
```

```python
df_clientes.info()
```

```python
#columna cuentas diarias
df_clientes['Cuentas_Diarias']= df_clientes['account.Charges.Monthly'] / 30
df_clientes
```

```python
#renombro columnas
df_clientes = df_clientes.rename(columns={
                                          "customer.gender":"Gender",
                                          "customer.SeniorCitizen":">65",
                                          "customer.Partner":"Partner",
                                          "customer.Dependents":"Dependents",
                                          "customer.tenure":"Tenure",
                                          "phone.PhoneService":"PhoneService",
                                          "phone.MultipleLines":"MultipleLines",
                                          "internet.InternetService":"InternetService",
                                          "internet.DeviceProtection":"DeviceProtection",
                                          "internet.TechSupport":"TechSupport",
                                          "internet.StreamingTV":"StreamingTV",
                                          "internet.StreamingMovies":"StreamingMovies",
                                          "account.Contract":"Contract",
                                          "account.PaperlessBilling":"PaperlessBilling",
                                          "account.PaymentMethod":"PaymentMethod",
                                          "account.Charges.Monthly":"ChargesMonthly",
                                          "account.Charges.Total":"ChargesTotal",
                                          "Cuentas_Diarias":"ChargesDaily"
})
```

```python
df_clientes
```

```python
columnas_binarias = [
                      "Churn",
                      "Partner",
                      "Dependents",
                      "PhoneService",
                      "PaperlessBilling",
                      ]

for col in columnas_binarias:
    df_clientes[col] = df_clientes[col].map({"Yes":1, "No":0})
```

```python
df_clientes.sample(5)
```

```python
df_clientes.to_csv("datos_tratados.csv", index=False)
```

#📊 Carga y análisis

```python
#Análisis Descriptivo
df_clientes.describe()
```

```python
#Distribución de Evasión
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(7,5))

sns.set_style("whitegrid")

ax = sns.countplot(
    x="Churn",
    data=df_clientes,
    palette="Set2"
)

plt.title("Distribución de Cancelación de Clientes (Churn)", fontsize=14, fontweight="bold")
plt.xlabel("Estado del Cliente", fontsize=12)
plt.ylabel("Cantidad de Clientes", fontsize=12)

# Agregar etiquetas con valores
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=11)

plt.tight_layout()
plt.show()
```

##**Recuento de evasión por variables categóricas**

```python
plt.figure(figsize=(7,5))

sns.set_style("whitegrid")

ax = sns.countplot(
    x="Gender",
    hue="Churn",
    data=df_clientes,
    palette="Set2"
)

plt.title("Distribución de Churn por Género", fontsize=14, fontweight="bold")
plt.xlabel("Género", fontsize=12)
plt.ylabel("Número de Clientes", fontsize=12)

# Etiquetas en las barras
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}',
                    (p.get_x() + p.get_width()/2., height),
                    ha='center', va='bottom', fontsize=10)

plt.legend(title="Churn")
plt.tight_layout()

plt.show()
```

```python
#Tipo de contrato
plt.figure(figsize=(8,5))

sns.set_style("whitegrid")

ax = sns.countplot(
    x="Contract",
    hue="Churn",
    data=df_clientes,
    palette="Set2",
    order=["Month-to-month", "One year", "Two year"]
)

plt.title("Cancelación de Clientes por Tipo de Contrato", fontsize=14, fontweight="bold")
plt.xlabel("Tipo de Contrato", fontsize=12)
plt.ylabel("Cantidad de Clientes", fontsize=12)

# Etiquetas de valores
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}',
                    (p.get_x() + p.get_width()/2., height),
                    ha='center', va='bottom',
                    fontsize=10)

plt.legend(title="Churn")
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

```python
#Método de pago
plt.figure(figsize=(9,5))

sns.set_style("whitegrid")

ax = sns.countplot(
    x="PaymentMethod",
    hue="Churn",
    data=df_clientes,
    palette="Set2"
)

plt.title("Churn por Método de Pago", fontsize=14, fontweight="bold")
plt.xlabel("Método de Pago", fontsize=12)
plt.ylabel("Cantidad de Clientes", fontsize=12)

# Etiquetas con valores
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}',
                    (p.get_x() + p.get_width()/2., height),
                    ha='center', va='bottom', fontsize=10)

plt.xticks(rotation=35)
plt.legend(title="Churn")

plt.tight_layout()
plt.show()
```

```python
#Tipo de servicio
plt.figure(figsize=(8,5))

sns.set_style("whitegrid")

ax = sns.countplot(
    x="InternetService",
    hue="Churn",
    data=df_clientes,
    palette="Set2"
)

plt.title("Churn por Tipo de Servicio de Internet", fontsize=14, fontweight="bold")
plt.xlabel("Tipo de Internet", fontsize=12)
plt.ylabel("Cantidad de Clientes", fontsize=12)

# Agregar etiquetas de valores
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}',
                    (p.get_x() + p.get_width()/2., height),
                    ha='center', va='bottom', fontsize=10)

plt.legend(title="Churn")
plt.tight_layout()

plt.show()
```

##**Conteo de evasión por variables numéricas**

```python
##Conteo de evasión por variables numéricas
plt.figure(figsize=(7,5))

sns.set_style("whitegrid")

ax = sns.boxplot(
    x="Churn",
    y="Tenure",
    data=df_clientes,
    palette="Set2"
)

# Agregar puntos de distribución
sns.stripplot(
    x="Churn",
    y="Tenure",
    data=df_clientes,
    color="gray",
    alpha=0.3
)

plt.title("Tiempo como Cliente vs Churn", fontsize=14, fontweight="bold")
plt.xlabel("Estado de Cancelación", fontsize=12)
plt.ylabel("Meses como Cliente (Tenure)", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.boxplot(x="Churn", y="ChargesMonthly", data=df_clientes, palette="Set2")

plt.title("Cargos Mensuales vs Churn", fontweight="bold")
plt.xlabel("Churn")
plt.ylabel("Cargos Mensuales")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.set_style("whitegrid")

sns.histplot(
    data=df_clientes,
    x="Tenure",
    hue="Churn",
    bins=30,
    kde=True,
    alpha=0.6
)

plt.title("Distribución del Tiempo como Cliente según Churn", fontsize=14, fontweight="bold")
plt.xlabel("Meses como Cliente (Tenure)", fontsize=12)
plt.ylabel("Cantidad de Clientes", fontsize=12)

plt.tight_layout()
plt.show()
```

#**Informe de Análisis de Evasión de Clientes (Churn) – Telecom X**

1. **Introducción**

    La empresa Telecom X enfrenta una alta tasa de cancelación de clientes, fenómeno conocido como Churn o evasión de clientes. Este problema representa un desafío importante para la empresa, ya que la pérdida constante de clientes impacta directamente en los ingresos y en el crecimiento del negocio.
    El objetivo de este análisis es explorar y comprender los factores que influyen en la cancelación del servicio por parte de los clientes. A partir del análisis de los datos disponibles, se busca identificar patrones y características comunes entre los clientes que abandonan la empresa.
    Los resultados obtenidos servirán como base para que el equipo de Data Science pueda desarrollar modelos predictivos de churn y diseñar estrategias efectivas de retención de clientes.

2. **Limpieza y Tratamiento de Datos**

    Para preparar los datos para el análisis, se realizaron varios pasos de procesamiento y limpieza:

    **Importación de datos**
    Los datos fueron obtenidos a través de una API en formato JSON y posteriormente convertidos a un DataFrame de Pandas para facilitar su manipulación y análisis.

    **Exploración inicial**
    Se revisó la estructura del dataset mediante:
    * Identificación de columnas
    *  Verificación de tipos de datos
    * Revisión del diccionario de datos para comprender el significado de cada variable

    **Limpieza de datos**
    Durante esta etapa se realizaron las siguientes acciones:
    * Identificación de valores nulos o vacíos
    * Conversión de variables a tipos de datos adecuados
    * Eliminación de registros duplicados
    * Corrección de inconsistencias en algunas columnas


    **Creación de nuevas variables**
    Se creó la columna Cuentas_Diarias, calculada a partir de la facturación mensual:

    df_clientes['Cuentas_Diarias']= df_clientes['account.Charges.Monthly'] / 30

    Esta variable permite analizar el comportamiento de facturación de los clientes de manera más detallada.

    **Transformación de datos**
    Para facilitar el análisis y futuros modelos analíticos, se realizaron transformaciones como:
    * Conversión de variables categóricas como Yes/No a valores binarios 1/0
    * Renombramiento de columnas para mejorar la claridad del dataset


3. **Análisis Exploratorio de Datos (EDA)**  
Se realizaron diferentes análisis para comprender mejor el comportamiento de los clientes.

    **Distribución de Churn**
    Se analizaron las proporciones de clientes que permanecen en la empresa frente a aquellos que cancelaron el servicio.

    image.png

    Las visualizaciones mostraron que, aunque la mayoría de los clientes continúa con el servicio, existe un porcentaje significativo de cancelaciones, lo que evidencia la importancia de estudiar este fenómeno.

    **Análisis de variables categóricas**
    Se exploró cómo la evasión se distribuye según variables categóricas como:
      * Género
      * Tipo de contrato

      image.png
      * Método de pago
      * Tipo de servicio de internet

    Los gráficos revelaron que el tipo de contrato es uno de los factores más asociados con la cancelación, especialmente en clientes con contratos mensuales.

    **Análisis de variables numéricas**
    También se analizaron variables numéricas relevantes como:
      * Tiempo como cliente (tenure)

      image.png
      * Cargos mensuales
      * Total

    Los gráficos de distribución y boxplots permitieron identificar diferencias claras entre los clientes que cancelaron y los que permanecen.


4. **Conclusiones e Insights**
A partir del análisis exploratorio se identificaron varios patrones importantes relacionados con la evasión de clientes:

    * Los clientes con contratos mensuales presentan una mayor tasa de cancelación en comparación con aquellos con contratos de largo plazo.

    * Los clientes con menor tiempo en la empresa tienen mayor probabilidad de abandonar el servicio.

    * El  método de pago electronic check presenta mayor proporción de churn.

  Estos hallazgos sugieren que tanto factores contractuales como económicos influyen en la decisión de los clientes de cancelar el servicio.
