#📌 Extracción

```python
url = "https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json"
```

```python
import pandas as pd
import requests
import json
import warnings
warnings.filterwarnings('ignore')
```

```python
datos_clientes = requests.get(url)
```

```python
resultado= json.loads(datos_clientes.text)
pd.DataFrame(resultado)
```

```python
df_clientes = pd.json_normalize(resultado)
df_clientes
```

## Diccionario de datos

* **customerID:** número de identificación único de cada cliente
* **Churn:** si el cliente dejó o no la empresa
* **gender:** género (masculino y femenino)
* **SeniorCitizen:** información sobre si un cliente tiene o no una edad igual o mayor a 65 años
* **Partner:** si el cliente tiene o no una pareja
* **Dependents:** si el cliente tiene o no dependientes
* **tenure:** meses de contrato del cliente
* **PhoneService:** suscripción al servicio telefónico
* **MultipleLines:** suscripción a más de una línea telefónica
* **InternetService:** suscripción a un proveedor de internet
* **OnlineSecurity:** suscripción adicional de seguridad en línea
* **OnlineBackup:** suscripción adicional de respaldo en línea
* **DeviceProtection:** suscripción adicional de protección del dispositivo
* **TechSupport:** suscripción adicional de soporte técnico, menor tiempo de espera
* **StreamingTV:** suscripción de televisión por cable
* **StreamingMovies:** suscripción de streaming de películas
* **Contract:** tipo de contrato
* **PaperlessBilling:** si el cliente prefiere recibir la factura en línea
* **PaymentMethod:** forma de pago
* **Charges.Monthly:** total de todos los servicios del cliente por mes
* **Charges.Total:** total gastado por el cliente

#🔧 Transformación

```python
#duplicados
df_clientes.duplicated().sum()
```

```python
#datos unicos
df_clientes.nunique()
```

```python
for col in df_clientes.columns:
  print(f'{col}: {df_clientes[col].nunique()}')
  if df_clientes[col].nunique() < 50:
    print(df_clientes[col].unique())
    print('-' * 50)
```

```python
#nulos
df_clientes.isnull().sum()
```

```python
#celdas vacias
(df_clientes == "").sum()
```

```python
#elimio filas vacias de churn
df_clientes= df_clientes[df_clientes['Churn'].str.strip() != '']
```

```python
#verifico celdas vacias
(df_clientes == "").sum()
```

```python
#celdas con cadenas vacias
(df_clientes == " ").sum()
```

```python
#tranformo a float Charges y elimino vacios
df_clientes["account.Charges.Total"] = pd.to_numeric(df_clientes["account.Charges.Total"], errors="coerce")
df_clientes = df_clientes.dropna(subset=["account.Charges.Total"])
df_clientes
```

```python
#verifico
(df_clientes == " ").sum()
```

```python
df_clientes[df_clientes["account.Charges.Total"].isna()]
```

```python
df_clientes.info()
```

```python
#columna cuentas diarias
df_clientes['Cuentas_Diarias']= df_clientes['account.Charges.Monthly'] / 30
df_clientes
```

```python
#renombro columnas
df_clientes = df_clientes.rename(columns={
                                          "customer.gender":"Gender",
                                          "customer.SeniorCitizen":">65",
                                          "customer.Partner":"Partner",
                                          "customer.Dependents":"Dependents",
                                          "customer.tenure":"Tenure",
                                          "phone.PhoneService":"PhoneService",
                                          "phone.MultipleLines":"MultipleLines",
                                          "internet.InternetService":"InternetService",
                                          "internet.DeviceProtection":"DeviceProtection",
                                          "internet.TechSupport":"TechSupport",
                                          "internet.StreamingTV":"StreamingTV",
                                          "internet.StreamingMovies":"StreamingMovies",
                                          "account.Contract":"Contract",
                                          "account.PaperlessBilling":"PaperlessBilling",
                                          "account.PaymentMethod":"PaymentMethod",
                                          "account.Charges.Monthly":"ChargesMonthly",
                                          "account.Charges.Total":"ChargesTotal",
                                          "Cuentas_Diarias":"ChargesDaily"
})
```

```python
df_clientes
```

```python
columnas_binarias = [
                      "Churn",
                      "Partner",
                      "Dependents",
                      "PhoneService",
                      "PaperlessBilling",
                      ]

for col in columnas_binarias:
    df_clientes[col] = df_clientes[col].map({"Yes":1, "No":0})
```

#📊 Carga y análisis

```python
#Análisis Descriptivo
df_clientes.describe()
```

```python
#Distribución de Evasión
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="Churn", data=df_clientes)

plt.title("Distribución de Cancelación de Clientes (Churn)", fontweight="bold")
plt.xlabel("Estado del Cliente")
plt.ylabel("Cantidad de Clientes")

plt.show()
```

##**Recuento de evasión por variables categóricas**

```python

#churn gender
plt.figure(figsize=(6,4))

sns.countplot(x="Gender", hue="Churn", data=df_clientes)

plt.title("Churn for Gender", fontweight="bold")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")

plt.show()
```

```python
#Tipo de contrato
plt.figure(figsize=(8,4))

sns.countplot(x="Contract", hue="Churn", data=df_clientes)

plt.title("Churn por Tipo de Contrato", fontweight="bold")
plt.xlabel("Tipo de Contrato")
plt.ylabel("Cantidad de Clientes")

plt.xticks(rotation=45)

plt.show()
```

```python
#Método de pago
plt.figure(figsize=(8,4))

sns.countplot(x="PaymentMethod", hue="Churn", data=df_clientes)

plt.title("Churn por Método de Pago", fontweight="bold")
plt.xlabel("Método de Pago")
plt.ylabel("Cantidad de Clientes")

plt.xticks(rotation=45)

plt.show()
```

```python
#Tipo de servicio
plt.figure(figsize=(7,4))

sns.countplot(x="InternetService", hue="Churn", data=df_clientes)

plt.title("Churn por Tipo de Servicio de Internet", fontweight="bold")
plt.xlabel("Tipo de Internet")
plt.ylabel("Cantidad de Clientes")

plt.show()
```

```python
#TV
plt.figure(figsize=(6,4))

sns.countplot(x="StreamingTV", hue="Churn", data=df_clientes)

plt.title("Churn según Servicio de Streaming TV", fontweight="bold")
plt.xlabel("Streaming TV")
plt.ylabel("Cantidad de Clientes")

plt.show()
```

```python
pd.crosstab(df_clientes["Contract"], df_clientes["Churn"], normalize="index") * 100
```

##**Conteo de evasión por variables numéricas**

```python
##Conteo de evasión por variables numéricas
plt.figure(figsize=(6,4))

sns.boxplot(x="Churn", y="Tenure", data=df_clientes)

plt.title("Tiempo como Cliente vs Churn", fontweight="bold")
plt.xlabel("Churn")
plt.ylabel("Meses como Cliente")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.boxplot(x="Churn", y="ChargesMonthly", data=df_clientes)

plt.title("Cargos Mensuales vs Churn", fontweight="bold")
plt.xlabel("Churn")
plt.ylabel("Cargos Mensuales")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.boxplot(x="Churn", y="ChargesTotal", data=df_clientes)

plt.title("Total Gastado vs Churn", fontweight="bold")
plt.xlabel("Churn")
plt.ylabel("Total Gastado")

plt.show()
```

```python
plt.figure(figsize=(7,4))

sns.histplot(data=df_clientes, x="ChargesMonthly", hue="Churn", bins=30, kde=True)

plt.title("Distribución de Cargos Mensuales según Churn", fontweight="bold")

plt.show()
```
